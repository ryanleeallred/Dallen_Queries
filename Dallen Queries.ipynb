## Replace multiple whitespace with single whitespace in DrugLabelName

```python
df['DrugLabelName'] = df['DrugLabelName'].astype(str)
df['DrugLabelName'] = df['DrugLabelName'].apply(lambda drug: ' '.join(drug.split()))
```

## Filter out rows where ingredient cost is <= 0

```python
df = df[df['IngredientCost'] > 0]
print(df.shape)
df.head()
```

## Filter out rows where quantity <= 0

```python
df = df[df['Quantity'] > 0]
print(df.shape)
df.head()
```

## Groupby NDC Code
## Create Dataframe of the Groupby to get Counts of NDC Codes

```python
grouped = df.groupby(by='NDCCode', as_index=False).size()
grouped.name = 'count'
grouped = grouped.reset_index()
grouped.head()
```

## Find Average of IngredientCost by NCDCode and add to dataframe

```python
average_ingredient_cost_list = df.groupby('NDCCode')['IngredientCost'].mean().tolist()
average_ingredient_cost_series = pd.Series(average_ingredient_cost_list)
grouped['AverageIngredientCost'] = average_ingredient_cost_series.values
grouped.head()
```

## Find Average of Quantity by NCDCode and add to dataframe

```python
average_quantity_list = df.groupby('NDCCode')['Quantity'].mean().tolist()
average_quantity_series = pd.Series(average_quantity_list)
grouped['AverageQuantity'] = average_quantity_series.values
grouped.head()
```

## Sort by NDC Code Counts

```python
sorted = grouped.sort_values(by=['count'], ascending=False)
sorted.head()
```

## Groupby DrugLabelName and sort to check if NDC Code and DrugLabelName are 1:1

```python
names = df.groupby(by='DrugLabelName', as_index=False).size()
names.name = 'count'
names = names.reset_index()
names.head()
```

```python
sorted_names = names.sort_values(by=['count'], ascending=False)
sorted_names.head()
```
